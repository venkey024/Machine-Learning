**1.Nike Analysis**


```python
df_nike = df[df['brand'] == "Nike"]
df_nike
```

```python
df_nike.isna().sum()
```

```python
median = df['sold_qty'].median()
df_nike['sold_qty'] = df_nike['sold_qty'].fillna(median)
df_nike
```

```python
df_nike.isna().sum()
```

```python
df.isna().sum()
```

**2.Addidas Analysis**

```python
df_adid = df[df['brand'] == 'Adidas']
df_adid
```

```python
df_adid_median = df_adid['sold_qty'].median()
df_adid['sold_qty'] = df_adid['sold_qty'].replace(df_adid['sold_qty'].max(),df_adid_median)
```

```python
df_adid['sold_qty'].max()
df_adid.isna().sum()
```

```python
df.isna().sum()
```

```python
df = pd.concat([df_nike,df_adid],ignore_index=True)
df
```

```python
plt.plot(df_nike['date'][0:5],df_nike['sold_qty'][0:5])
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by brand and sum sold quantity
brand_sales = df.groupby('brand')['sold_qty'].sum()

# Plotting
brand_sales.plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Total Sold Quantity by Brand')
plt.xlabel('Brand')
plt.ylabel('Total Sold Quantity')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=df, x='brand', y='sold_qty', palette='Set2')
plt.title('Distribution of Sold Quantity by Brand')
plt.xlabel('Brand')
plt.ylabel('Sold Quantity')
plt.grid(True)
plt.show()
```
