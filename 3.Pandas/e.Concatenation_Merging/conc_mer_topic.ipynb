```python
import pandas as pd
```

```python
india_weather = pd.DataFrame({
    'city' : ['Mumbai','Hyderabad','Nellore'],
    'temparature':[30,34,29],
    'humidity':[40,52,50]
})
```

```python
us_weather = pd.DataFrame({
    'city' : ['Newyork','Chicacgo','amit'],
    'temparature':[20,44,29],
    'humidity':[30,32,23]
})
```

```python
df = pd.concat([india_weather,us_weather])
df
```

```python
# we can ignore index directly.
df = pd.concat([india_weather,us_weather],ignore_index = True)
df
```

```python
# we can pass keys to df
df = pd.concat([india_weather,us_weather],keys=['India','US'])
df
```

```python
# we can caoncatenate columns wise - passing with axis=1
df = pd.concat([india_weather,us_weather],axis=1)
df
```
