```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
line_df = pd.read_excel("linechart.xlsx")
line_df
```

```python
plt.plot(line_df['Quarter'],line_df['Dishwasher'])
plt.show()
```

```python
# if we mentioned figure size then it will cover everything space
plt.figure(figsize=(14,4))
plt.plot(line_df['Quarter'],line_df['Dishwasher'])
plt.show()
```

```python
# we can customize our colors
plt.figure(figsize=(14,4))
plt.plot(line_df['Quarter'],line_df['Dishwasher'],color="red")
plt.plot(line_df['Quarter'],line_df['Fridge'])
plt.plot(line_df['Quarter'],line_df['Washing Machine'])
plt.title('Summary of Quarter wise revenues')
plt.xlabel('Quarters')
plt.ylabel('Sum of Revenues')
plt.show()
```

```python
total_sales = line_df[['Fridge','Dishwasher','Washing Machine']].sum()
total_sales.index
```

```python
plt.pie(total_sales,labels = total_sales.index,autopct='%1.2f%%')
plt.show()
```

```python
#autopct = %1.1f%' is for percentage in the pie chart
# explode is for extract labels from chart
# startangle=True means we are setting angle for rotating
# explode=(indexes of data frame) --> total_sales.index = Index(['Fridge', 'Dishwasher', 'Washing Machine'], dtype='object')

plt.pie(total_sales,labels = total_sales.index,autopct='%1.2f%%',explode=(0.05,0,0),startangle=140,shadow=True)
plt.show()
```
