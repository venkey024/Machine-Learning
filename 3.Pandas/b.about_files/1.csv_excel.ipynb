```python
import pandas as pd
```

```python
df =  pd.read_csv('stock_data.csv')
df
```

```python
# we can skip the rows directly for column name
df =  pd.read_csv('stock_data.csv',skiprows=1)
df.head()
```

```python
# we have another option fro setting header
df =  pd.read_csv('stock_data.csv',header=1)
df.head()
```

```python
# we can give custom columns
df =  pd.read_csv('stock_data.csv',skiprows=1,names=['names','eps','revenue','price','people'])
df.head()
```

```python
# if we want to replace somevalues with NaN

df =  pd.read_csv('stock_data.csv',skiprows=1,names=['names','eps','revenue','price','people'],na_values={'eps':['not available']})
df.head()
```
