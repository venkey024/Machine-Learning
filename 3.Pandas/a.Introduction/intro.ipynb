```python
# adding new column based on new release_year.
# using for loop

ages = []
for i in df['release_year']:
    x = 2025-i
    ages.append(x)
df['age'] = ages
df.head()
```

```python
# usig short normal condition
df['ages_n'] = 2025-df['release_year']
df.head()
```

```python
# i want profit.

df['profit'] = df['revenue'] - df['budget']
df[df['language'] == 'Telugu']
```

```python
df['profit_n'] = df.apply(lambda x : x['revenue'] - x['budget'],axis =1)
df.head()
```

```python
# axis = 1, for row wise operation  ---->  example profit = revenue - budget // for every row


df['n_release'] = df.apply(lambda x: 2025 - x['age'], axis=1)
df.head()
```
