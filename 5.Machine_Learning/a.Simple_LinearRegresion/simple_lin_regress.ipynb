**Simple Linear Regression**


```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("home_prices.csv")
df
```

```python
plt.figure(figsize=(14,7))
sns.scatterplot(x=df['area_sqr_ft'],y=df['price_lakhs'])
plt.show()
```

```python
# Importing the LinearRegression class from sklearn's linear_model module
from sklearn.linear_model import LinearRegression

# Creating an instance (object) of the LinearRegression model
model = LinearRegression()

# Training (fitting) the model using the dataset
# df[['area_sqr_ft']] is the feature (independent variable) – area of the house in square feet
# df['price_lakhs'] is the target (dependent variable) – price of the house in lakhs
model.fit(df[['area_sqr_ft']], df['price_lakhs'])
```

## Points

**1.** `model = LinearRegression()`  
Creates a linear regression model that will try to learn the relationship between `area_sqr_ft` and `price_lakhs`.

**2.** `model.fit(...)`  
This is the training step. The model learns the best-fit line between `area_sqr_ft` and `price_lakhs` by minimizing the error between actual and predicted prices.

**3.** `model.coef_`  
This returns the **coefficient** (slope) of the best-fit line. It represents how much the `price_lakhs` is expected to increase for every 1 unit increase in `area_sqr_ft`.

**4.** `model.intercept_`  
This returns the **intercept** of the line. It is the predicted value of `price_lakhs` when `area_sqr_ft` is 0.


```python
model.predict([[1500]])
```

```python
model.coef_,model.intercept_
```
